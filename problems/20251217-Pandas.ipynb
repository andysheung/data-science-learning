```python
#extract number in rupee
df["car_prices_in_rupee2"]= df["car_prices_in_rupee"].str.extract(r'(\d+(?:\.\d+)?)')[0].astype(float)
df["car_prices_in_rupee2"]=df["car_prices_in_rupee2"]*100000
df["car_prices_in_rupee2"]= df["car_prices_in_rupee2"].round(0).astype(int)
df["car_prices_in_rupee2"]
df
```

```python
df = df[['car_name','car_prices_in_rupee2','kms_driven','fuel_type','transmission','ownership','manufacture','engine','Seats']]
df
```

```python
df.rename(columns={"car_prices_in_rupee2":"car_prices_in_rupee"}, inplace=True)
df
```

```python
df['kms_driven'] = df['kms_driven'].str.replace(",","")
df['kms_driven'] = df['kms_driven'].str.extract(r'(\d+)')[0]
df
```

```python
df['Make'] = df["car_name"].str.extract(r'(\w+(?:\-\w+)?)')[0]
df['Model'] = df["car_name"].str.replace(r'^\w+(?:\-\w+)?',"",regex=True)
df['Model']
```
