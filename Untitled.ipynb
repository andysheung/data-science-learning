```python
from collections import defaultdict
class Solution:
    def removeDuplicates(self, nums: list[int]) -> int:
        check_dict=defaultdict(int)
        k = 0 #counter for first k elements appears at most twice in nums
        j = len(nums)-1 # 5
        i = 0
        while i <= j: # len=6, [0,1,2,3,4,5]
            print(f'i:{i},j:{j},k:{k},nums:{nums}')
            if check_dict[nums[i]] == 2:
                nums.pop(i)
                nums.append(0)
                j-=1
            else: 
                check_dict[nums[i]]+=1
                k+=1
                i+=1
            print(f'k:{k},nums:{nums}')
        return k
```

```python
nums =[1,1,1,2,2,3]
```

```python
sol = Solution()
sol.removeDuplicates(nums=nums)
```
